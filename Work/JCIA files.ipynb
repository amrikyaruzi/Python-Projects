```python
#Nursing care plan
#BT (not a parameter, what files have? are complete?)
#STAT medications
#CPOE orders - doctors signature
```

```python
data2 = data1[(data1["DOCUMENTS"] == "Informed Consent") | (data1["DOCUMENTS"] == "Physician Initial Assessment") |
              (data1["DOCUMENTS"] == "Nursing Initial Assessment") | (data1["DOCUMENTS"] == "CPOE orders - Nurse's Signature") |
              (data1["DOCUMENTS"] == "CPOE orders - Doctor's Signature") |
              (data1["DOCUMENTS"] == "Multidisciplinary Patient and Family Education") | (data1["DOCUMENTS"] == "Procedure/ Operative notes") |
              (data1["DOCUMENTS"] == "Site Marking, Sign in, Time out & Sign out") | (data1["DOCUMENTS"] == "STAT medications") | 
              (data1["DOCUMENTS"] == "Multidisciplinary Patient and Family Education Form (Doctor's Part)") | 
              (data1["DOCUMENTS"] == "Multidisciplinary Patient and Family Education Form (Nurse's Part)")]
```

```python
data2["DEPT"].unique()
```

```python
data2 = data2[(data2["MET"] == "Y")|(data2["MET"] == "N")]
data2 = data2[(data2["DEPT"] == "Surgery")|
              (data2["DEPT"] == "Maternity")|
              (data2["DEPT"] == "Medicine")|
              (data2["DEPT"] == "Paediatrics")]
```

```python
index_list = ["AKNO", "DEPT", "DOCTOR", "Month"]
data3 = data2.pivot_table(index = index_list, columns = "DOCUMENTS", values = "MET", aggfunc= np.sum)
```

```python
data3.reset_index()
```

```python
data3.fillna("N/A", inplace = True)
```

```python
data3['Use for JCIA'] = np.where((
                                  ((data3['Informed Consent']=='N/A') | (data3['Informed Consent'] == 'Y')) &
                                  ((data3["CPOE orders - Nurse's Signature"] == "Y") | (data3["CPOE orders - Nurse's Signature"] == "N/A"))&
                                  ((data3["STAT medications"] == "Y") | (data3["STAT medications"] == "N/A")) &
                                  ((data3["Site Marking, Sign in, Time out & Sign out"] == "Y") | (data3["Site Marking, Sign in, Time out & Sign out"] == "N/A")) &
                                  ((data3["Procedure/ Operative notes"] == "Y") | (data3["Procedure/ Operative notes"] == "N/A")) &
                                  ((data3["Multidisciplinary Patient and Family Education Form (Doctor's Part)"] == "Y") |
                                   (data3["Multidisciplinary Patient and Family Education Form (Doctor's Part)"] == "N/A")) &
                                  ((data3["Multidisciplinary Patient and Family Education Form (Nurse's Part)"] == "Y") |
                                   (data3["Multidisciplinary Patient and Family Education Form (Nurse's Part)"] == "N/A")) & 
                                  ((data3["Multidisciplinary Patient and Family Education"] == "Y") |
                                   (data3["Multidisciplinary Patient and Family Education"] == "N/A")) &
                                  ((data3["CPOE orders - Doctor's Signature"] == "Y") | (data3["CPOE orders - Doctor's Signature"] == "N/A"))
                                  ), 'Maybe', 'No')
```

```python
#data3['Use for JCIA'] = np.where((((data3['Informed Consent']=='N/A') | (data3['Informed Consent'] == 'Y'))), 'Yes', 'No')
```

```python
data3 = data3.reset_index()
data3 = data3[data3["Month"].isin(["Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])]
data3.shape
```

```python
new_order = ["Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
data3["Month"] = pd.Categorical(data3["Month"], categories=new_order, ordered=True)
data3 = data3.sort_values(by="Month", ascending=True).reset_index()

#pd.to_datetime(data3["Month"], format="%b", errors="coerce").dt.month
```
